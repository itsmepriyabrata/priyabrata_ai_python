```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import time

# Generate random data
np.random.seed(0)
x = np.arange(0, 100, 1)
y = np.random.randn(100)

# Initialize the model
model = LinearRegression()


plt.ion()
fig, ax = plt.subplots()
ax.plot(x, y, 'b-', label='Random Data')


for i in range(100):

    model.fit(x[:i+1].reshape(-1, 1), y[:i+1])


    next_x = np.array([[i+1]])
    next_y = model.predict(next_x)


    ax.plot(i+1, next_y, 'r.', markersize=10, label='Prediction' if i == 0 else None)


    plt.pause(1)  # Pause for 1 second
    plt.draw()
```
