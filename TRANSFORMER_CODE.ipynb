```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained GPT-2 model and tokenizer
model_name = "gpt2-medium"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

prompt = "Discuss the importance of education in Odisha and its impact on society."

input_ids = tokenizer.encode(prompt, return_tensors="pt")

output = model.generate(input_ids, max_length=200, num_return_sequences=1, temperature=0.7, top_k=50)

generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print("Generated Text:")
print(generated_text)
```

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer

# Load the T5 model and tokenizer
model_name = "t5-small"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

prompts = [
    "Tell me about the tourist attractions in Odisha.",
    "Can you recommend some famous beaches in Odisha?",
    "What are the popular festivals celebrated in Odisha?",
    "Provide information about the wildlife sanctuaries in Odisha.",
]

for prompt in prompts:
    input_ids = tokenizer.encode(prompt, return_tensors="pt", max_length=512, truncation=True)

    output = model.generate(input_ids, max_length=200, num_return_sequences=1, early_stopping=True)

    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)

    # Print the generated text
    print("Prompt:", prompt)
    print("Generated Text:", generated_text)
    print()
```
