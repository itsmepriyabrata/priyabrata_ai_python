```python
import matplotlib.pyplot as plt

# Monthly profit data (replace with your actual data)
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
profits = [10000, 12000, 11000, 13000, 12500, 14000, 15000, 16000, 15500, 16500, 17000, 18000]

total_profit = sum(profits)

plt.figure(figsize=(10, 6))

plt.plot(months, profits, marker='o', color='b', linestyle='-', linewidth=2, markersize=8)

plt.title('Total Profit Over Months', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Profit', fontsize=14)

plt.text(0.5, 0.9, f'Total Profit: ${total_profit}', transform=plt.gca().transAxes, fontsize=12)

plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Replace this with your actual data
total_profits = [10000, 12000, 11000, 13000, 12500, 14000, 15000, 16000, 15500, 16500, 17000, 18000]

months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

plt.figure(figsize=(10, 6))

plt.plot(months, total_profits, marker='o', linestyle='-')

plt.title('Total Profit Over Months')
plt.xlabel('Month')
plt.ylabel('Total Profit')

plt.grid(True)

plt.xticks(rotation=45)

# Showing the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

product_sales_data = {
    'Product A': [100, 120, 110, 130, 125, 140, 150, 160, 155, 165, 170, 180],
    'Product B': [90, 110, 105, 115, 120, 130, 135, 140, 145, 150, 155, 160],
    'Product C': [80, 100, 95, 105, 110, 120, 125, 130, 135, 140, 145, 150]
}

months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Plotting
plt.figure(figsize=(10, 6))

for product, sales_data in product_sales_data.items():
    plt.plot(months, sales_data, marker='o', label=product)

plt.title('Product Sales Over Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()  # Add legend to show which line corresponds to which product

# Adding grid
plt.grid(True)

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

toothpaste_sales = [100, 120, 110, 130, 125, 140, 150, 160, 155, 165, 170, 180]

months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']


plt.figure(figsize=(10, 6))

plt.scatter(months, toothpaste_sales, color='blue', label='Toothpaste Sales')

plt.title('Toothpaste Sales Over Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()  # Add legend

plt.grid(True)

plt.xticks(rotation=45)

# Showing the plot
plt.tight_layout()
plt.show()
```
