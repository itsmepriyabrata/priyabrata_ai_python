```python
class ToDoList:
    def __init__(self):
        self.categories = {
            "Personal": [
                {"task": "Meditation for 10 minutes", "completed": False},
                {"task": "Drink 2 liters of water", "completed": False},
            ],
            "Professional": [
                {"task": "Complete Python project feature", "completed": False},
                {"task": "Attend team meeting", "completed": False},
            ],
            "Health & Fitness": [
                {"task": "Go for a 30-minute walk", "completed": False},
                {"task": "Prepare a healthy meal", "completed": False},
            ],
        }

    def display_menu(self):
        print("\n--- To-Do List Menu ---")
        print("1. View Tasks")
        print("2. Add a Task")
        print("3. Mark a Task as Completed")
        print("4. Delete a Task")
        print("5. Exit")

    def view_tasks(self):
        print("\n--- Your To-Do List ---")
        for category, tasks in self.categories.items():
            print(f"\n{category}:")
            if not tasks:
                print("  No tasks in this category.")
            else:
                for i, task in enumerate(tasks, start=1):
                    status = "✔" if task['completed'] else "✖"
                    print(f"  {i}. {task['task']} [{status}]")

    def add_task(self):
        category = input("Enter the category (Personal, Professional, Health & Fitness): ").strip()
        if category in self.categories:
            task_name = input("Enter the task: ").strip()
            if task_name:
                self.categories[category].append({'task': task_name, 'completed': False})
                print(f"Task '{task_name}' added to {category}!")
            else:
                print("Task cannot be empty!")
        else:
            print("Invalid category! Please choose from the listed categories.")

    def mark_task_completed(self):
        category = input("Enter the category (Personal, Professional, Health & Fitness): ").strip()
        if category in self.categories:
            self.view_category_tasks(category)
            try:
                task_num = int(input("Enter the task number to mark as completed: "))
                if 1 <= task_num <= len(self.categories[category]):
                    self.categories[category][task_num - 1]['completed'] = True
                    print(f"Task '{self.categories[category][task_num - 1]['task']}' marked as completed!")
                else:
                    print("Invalid task number!")
            except ValueError:
                print("Please enter a valid number!")
        else:
            print("Invalid category!")

    def delete_task(self):
        category = input("Enter the category (Personal, Professional, Health & Fitness): ").strip()
        if category in self.categories:
            self.view_category_tasks(category)
            try:
                task_num = int(input("Enter the task number to delete: "))
                if 1 <= task_num <= len(self.categories[category]):
                    removed_task = self.categories[category].pop(task_num - 1)
                    print(f"Task '{removed_task['task']}' deleted from {category}!")
                else:
                    print("Invalid task number!")
            except ValueError:
                print("Please enter a valid number!")
        else:
            print("Invalid category!")

    def view_category_tasks(self, category):
        print(f"\n{category} Tasks:")
        tasks = self.categories[category]
        if not tasks:
            print("  No tasks in this category.")
        else:
            for i, task in enumerate(tasks, start=1):
                status = "✔" if task['completed'] else "✖"
                print(f"  {i}. {task['task']} [{status}]")

    def run(self):
        while True:
            self.display_menu()
            choice = input("Choose an option (1-5): ").strip()
            if choice == '1':
                self.view_tasks()
            elif choice == '2':
                self.add_task()
            elif choice == '3':
                self.mark_task_completed()
            elif choice == '4':
                self.delete_task()
            elif choice == '5':
                print("Exiting the To-Do List Application. Goodbye!")
                break
            else:
                print("Invalid choice! Please select a valid option.")

if __name__ == "__main__":
    app = ToDoList()
    app.run()
```
