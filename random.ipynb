```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import random
from IPython.display import HTML
import numpy as np
```

```python
class RandomNumberGenerator:
    def __init__(self):
        self.fig, self.ax = plt.subplots()
        self.x_data, self.y_data = [], []
        self.line, = self.ax.plot([], [], lw=2)
        self.animation = FuncAnimation(self.fig, self.update, frames=np.arange(0, 10), interval=1000)
        self.animation.save('random_number_animation.gif', writer='imagemagick', fps=1)

    def update(self, frame):
        random_number = random.randint(0, 100)
        self.x_data.append(frame)
        self.y_data.append(random_number)
        self.line.set_data(self.x_data, self.y_data)
        self.ax.relim()
        self.ax.autoscale_view()
        return self.line,

    def show_graph(self):
        return HTML('<img src="random_number_animation.gif">')

if __name__ == "__main__":
    rng = RandomNumberGenerator()
    rng.show_graph()
```
