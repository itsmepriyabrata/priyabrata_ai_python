```python
import networkx as nx
import random

def generate_world_model(num_nodes):
    G = nx.Graph()
    nodes = list(range(1, num_nodes + 1))
    G.add_nodes_from(nodes)

    for node in nodes:
        edges = random.sample(nodes, 3)
        G.add_edge(node, edges[0], weight=-10)
        G.add_edge(node, edges[1], weight=0)
        G.add_edge(node, edges[2], weight=10)

    return G

def visualize_graph(G):
    import matplotlib.pyplot as plt

    pos = nx.spring_layout(G)

    nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=500)
    labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edges(G, pos, edgelist=G.edges(), width=2, alpha=0.6, edge_color='grey')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_color='red')

    plt.axis('off')
    plt.show()

if __name__ == "__main__":
    num_countries = 196
    world_graph = generate_world_model(num_countries)
    visualize_graph(world_graph)

    degrees = dict(world_graph.degree())
    max_degree_nodes = [node for node, degree in degrees.items() if degree == max(degrees.values())]
    print(f"Nodes with the highest degree: {max_degree_nodes}, Degree: {max(degrees.values())}")
```
