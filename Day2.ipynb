```python
def are_all_numbers_unique(numbers):

    return len(set(numbers)) == len(numbers)

# Example usage:
sequence1 = [1, 2, 3, 4, 5]
sequence2 = [1, 2, 3, 4, 4]  # Duplicate 4

print(are_all_numbers_unique(sequence1))  # Output: True
print(are_all_numbers_unique(sequence2))  # Output: False
```

```python
from itertools import permutations

def generate_unique_strings(letters):

    unique_strings = set()  # Use a set to automatically remove duplicates

    # Generate all permutations of the letters
    for perm in permutations(letters):
        string = ''.join(perm)
        if len(string) == len(set(string)):  # Check if all characters are unique
            unique_strings.add(string)

    return list(unique_strings)

# Letters to be used
letters = 'aeioI'

# Generate and print all possible unique strings
unique_strings = generate_unique_strings(letters)
for string in unique_strings:
    print(string)
```

```python
def find_triplets(nums):

    nums.sort()  # Sort the array
    triplets = []

    n = len(nums)
    for i in range(n - 2):
        # Avoid duplicate triplets by skipping identical elements
        if i > 0 and nums[i] == nums[i - 1]:
            continue

        left = i + 1
        right = n - 1

        while left < right:
            total = nums[i] + nums[left] + nums[right]

            if total == 0:
                triplets.append([nums[i], nums[left], nums[right]])

                # Skip identical elements to avoid duplicate triplets
                while left < right and nums[left] == nums[left + 1]:
                    left += 1
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1

                left += 1
                right -= 1

            elif total < 0:
                left += 1
            else:
                right -= 1

    return triplets


# Example usage:
nums = [-1, 0, 1, 2, -1, -4]
result = find_triplets(nums)
print("Unique triplets whose sum is zero:")
for triplet in result:
    print(triplet)
```

```python
from itertools import combinations

def generate_digit_combinations():

    digits = range(10)  # Digits from 0 to 9
    digit_combinations = list(combinations(digits, 3))
    return digit_combinations

# Example usage:
combinations_of_3_digits = generate_digit_combinations()
print("Combinations of 3 digits:")
for combination in combinations_of_3_digits:
    print(combination)
```

```python
def count_word_frequency(text):
    """
    Count the frequency of each word in the given text.

    Parameters:
    text (str): The input text.

    Returns:
    dict: A dictionary where keys are words and values are their frequencies.
    """
    words = text.split()  # Split the text into words
    word_frequency = {}

    for word in words:
        # Remove punctuation marks if any
        word = word.strip('.,?!:;\'"()[]{}')
        # Convert the word to lowercase for case-insensitive counting
        word = word.lower()
        # Increment the frequency count for the word
        word_frequency[word] = word_frequency.get(word, 0) + 1

    return word_frequency

# Define a long text
long_text = """
Python is an easy to learn, powerful programming language. It has efficient high-level data structures and a simple but effective approach to object-oriented programming. Python's elegant syntax and dynamic typing, together with its interpreted nature, make it an ideal language for scripting and rapid application development in many areas on most platforms.
"""

# Print the long text
print("Long text:")
print(long_text)

# Count word frequency
frequency_dict = count_word_frequency(long_text)

# Print word frequency
print("\nWord frequency:")
for word, frequency in frequency_dict.items():
    print(f"{word}: {frequency}")
```
