```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained GPT-2 model and tokenizer
model_name = "gpt2-medium"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

prompt = "Education in Odisha has seen significant improvements over the years. "

input_ids = tokenizer.encode(prompt, return_tensors="pt", max_length=100, truncation=True)
output = model.generate(input_ids, max_length=200, num_return_sequences=1, temperature=0.9)

generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print("Generated Text:")
print(generated_text)
```

```python
import openai

# Set up your OpenAI API key
openai.api_key = 'api-key'

prompt = """
Odisha, often referred to as the soul of India, is a land of rich cultural heritage and stunning natural beauty.
From its magnificent temples and vibrant festivals to its pristine beaches and lush greenery, Odisha offers a unique
blend of tradition and modernity. Come explore the ancient wonders of Konark Sun Temple, Jagannath Temple, and
Lingaraja Temple, marvel at the intricate craftsmanship of Odishan art and handicrafts, and indulge in the
mouthwatering flavors of Odishan cuisine. Whether you're seeking spiritual enlightenment, adventure, or simply
a relaxing getaway, Odisha has something for everyone.

"""

response = openai.Completion.create(
  engine="text-davinci-002",
  prompt=prompt,
  max_tokens=150
)

print(response.choices[0].text.strip())
```
