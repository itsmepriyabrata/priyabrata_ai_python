```python
import matplotlib.pyplot as plt

toothpaste_sales = [100, 120, 110, 130, 125, 140, 150, 160, 155, 165, 170, 180]

months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Plotting
plt.figure(figsize=(10, 6))

plt.scatter(months, toothpaste_sales, color='blue', label='Toothpaste Sales')

plt.title('Toothpaste Sales Over Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()  # Add legend

plt.grid(True)

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

face_cream_sales = [2500, 2600, 2700, 2800, 2900, 3000, 3100, 3200, 3300, 3400, 3500, 3600]
facewash_sales = [1500, 1600, 1700, 1800, 1900, 2000, 2100, 2200, 2300, 2400, 2500, 2600]

# Months list for x-axis labels
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Plotting
plt.figure(figsize=(10, 6))

# Plotting bar chart for face cream sales data
plt.bar(months, face_cream_sales, color='blue', label='Face Cream Sales')

plt.bar(months, facewash_sales, color='orange', label='Facewash Sales', alpha=0.6)  # Using alpha to make bars semi-transparent

plt.title('Face Cream and Facewash Sales Over Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()  # Add legend

plt.xticks(rotation=45)

# Showing the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

bathing_soap_sales = [2000, 2100, 2200, 2300, 2400, 2500, 2600, 2700, 2800, 2900, 3000, 3100]

months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Plotting
plt.figure(figsize=(10, 6))

plt.bar(months, bathing_soap_sales, color='green', label='Bathing Soap Sales')

plt.title('Bathing Soap Sales Over Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()  # Add legend

plt.xticks(rotation=45)

plt.savefig('bathing_soap_sales_plot.png')

# Showing the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

total_profits = [10000, 12000, 11000, 13000, 12500, 14000, 15000, 16000, 15500, 16500, 17000, 18000]

# Plotting
plt.figure(figsize=(10, 6))

plt.hist(total_profits, bins=10, color='skyblue', edgecolor='black', alpha=0.7)

plt.title('Distribution of Total Profit')
plt.xlabel('Profit')
plt.ylabel('Frequency')

# Adding grid
plt.grid(True)

# Showing the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

product_sales = {
    'Product A': 50000,
    'Product B': 45000,
    'Product C': 60000,
    'Product D': 55000
}

plt.figure(figsize=(8, 8))

plt.pie(product_sales.values(), labels=product_sales.keys(), autopct='%1.1f%%', startangle=140)

plt.title('Total Sale Data for Last Year by Product')

# Showing the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```
